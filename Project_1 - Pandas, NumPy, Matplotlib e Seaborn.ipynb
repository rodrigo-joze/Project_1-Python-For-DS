### 1 - Importar bibliotecas Pandas e NumPy:


```python
import numpy as np
import pandas as pd
```

### 2 - Importar base de dados e inseri-lá em um dataframe:

```python
df = pd.read_csv('agricultural_raw_material.csv')
```

### 3 - Explorar o conjunto dos dados importados:

```python
# Apenas para visualizar um resumo da estrutura dos dados, utilizando o método .info:

df.info
```

```python
# Apenas para visualizar um resumo da estrutura dos dados, utilizando a função display():

display(df)
```

### 4 - Checar se há valores nulos e, se positivo, a quantidade em cada coluna/campo:

```python
# Utilizando os métodos .isnull() e .sum():

df.isnull().sum()
```

### 5 - Resolver os dados ausentes, incorretos e inválidos:

```python
# Substituir (replace) os valores por outros, utilizando a função .replace() e
# np.nan para inserir valor nulo

df = df.replace('%', '', regex = True)
df = df.replace(',', '', regex = True)
df = df.replace('-', '', regex = True)
df = df.replace('', np.nan)
df = df.replace('MAY90', np.nan)
```

```python
# Deletar os registros (rows/linhas) com valores nulos (NaN), utilizando a função .dropna():

df = df.dropna()
```

### 6 - Converter tipos de dados (data type), quando necessário.

```python
# Converter todos os tipos de dados (data type) para o tipo float, utilizando a função .astype():

# Primeiro, criar uma lista dos campos que quero converter:

lst = ["Coarse wool Price", "Coarse wool price % Change", "Copra Price", "Copra price % Change", "Cotton Price", "Cotton price % Change", "Fine wool Price", "Fine wool price % Change", "Hard log Price", "Hard log price % Change", "Hard sawnwood Price", "Hard sawnwood price % Change", "Hide Price", "Hide price % change", "Plywood Price", "Plywood price % Change", "Rubber Price", "Rubber price % Change", "Softlog Price", "Softlog price % Change", "Soft sawnwood Price", "Soft sawnwood price % Change", "Wood pulp Price", "Wood pulp price % Change"]
df[lst] = df[lst].astype("float")
df.dtypes
```

```python
# Mostrar os 5 primeiros registros do dataframe, utilizando .head():

df.head()
```

### 7 - Converter o campo 'Month', que é do tipo String, para data:



```python
# Conversão utilizando o método .to_datetime() e .set_index():

df.Month = pd.to_datetime(df.Month.str.upper(), format = '%b%y', yearfirst = False)
df = df.set_index('Month')
```

### 8 - Realizar a ANÁLISE EXPLORATÓRIA e VISUALIZAÇÃO

### 8.1 - Importar as bibliotecas (libraries) Matplotlib e Seaborn e definir os parâmetros para todos os gráficos:

```python
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt

%matplotlib inline
sns.set_style('darkgrid')
matplotlib.rcParams['font.size'] = 14
matplotlib.rcParams['figure.figsize'] = 9, 5
matplotlib.rcParams['figure.facecolor'] = '#00000000'
```

### 8.2 - Criar um MAPA DE CALOR que vai utilizar as cores conforme a CORRELAÇÃO das variáveis dos preços das matérias-primas:

```python
# Criar uma lista das matérias-primas (variáveis) para realizar a correlação e o gráfico:

raw_data = ['Coarse wool Price', 'Copra Price', 'Cotton Price', 'Fine wool Price', 'Hard log Price', 'Hard sawnwood Price', 'Hide Price', 'Plywood Price', 'Rubber Price', 'Softlog Price', 'Soft sawnwood Price', 'Wood pulp Price']

# Obter a matriz de correlação:

corrmat = df[raw_data].corr()

# Definir o tamanho do gráfico:

fig = plt.figure (figsize = (12, 9))

# Mascarar a parte superior do triângulo, pois a matriz é simétrica (repetitiva, exemplo: correlação de A e B é a mesma que B e A, para evitar repetição, utilizamos uma máscara para tratar essa duplicidade na hora de apresentar o gráfico):

mask = np.triu (np.ones_like (corrmat, dtype = bool))

# Aplicar a matriz acima ao mapa de calor:

sns.heatmap (corrmat, vmax = .8, mask = mask, square = True, annot = True)

# Mostrar o gráfico:
plt.show()
```

### Leitura do gráfico acima:

Quanto mais próximo de 1, maior a correlação. A maior correlação é entre Fine Wool (lã fina) e Coarse Wool (lã grossa), quando um aumento de valor o outro também aumenta e vice-versa.


### 8.3 - Criar um MAPA DE CALOR que vai utilizar as cores conforme a CORRELAÇÃO das variáveis das VARIAÇÕES dos preços das matérias-primas:

```python
# Criar uma lista das matérias-primas (variáveis) para realizar a correlação e o gráfico:

changelist = ['Coarse wool price % Change', 'Copra price % Change', 'Cotton price % Change', 'Fine wool price % Change', 'Hard log price % Change', 'Hard sawnwood price % Change', 'Hide price % change', 'Plywood price % Change', 'Rubber price % Change', 'Softlog price % Change', 'Soft sawnwood price % Change', 'Wood pulp price % Change']

# Obter a matriz de correlação:

corrMatrix = df[changelist].corr()

# Definir o tamanho do gráfico:

fig2 = plt.figure (figsize = (30, 15))

# Aplicar a matriz acima ao mapa de calor:

sns.heatmap (corrMatrix, annot = True)

# Mostrar o gráfico:
plt.show()
```

### Leitura do gráfico acima:

Há correlação entre os preços das matérias-primas, mas não há correlação forte (acima de ABS=0,7) entre as variações percentuais dos preços.

### 8.4 - Mapa de Séries Temporais (Time Series Map) da Lã Grossa por preço e por variação percentual do preço:

```python
# Usei subplots = True para separar os dois gráficos, pois trabalham com ordens de grandeza diferentes nos valores.

axes = df [["Coarse wool Price", "Coarse wool price % Change"]].plot (figsize = (11, 9), subplots = True, linewidth = 1)
```

### Leitura do gráfico acima:

O preço apresentou crescimento ao longo do tempo.

#### Q1: Descubra a variação normal do preço de cada matéria-prima.

```python
# Criar uma lista das matérias-primas (variáveis) para realizar o cálculo e plotar o gráfico:

changelist = ['Coarse wool price % Change', 'Copra price % Change', 'Cotton price % Change', 'Fine wool price % Change', 'Hard log price % Change', 'Hard sawnwood price % Change', 'Hide price % change', 'Plywood price % Change', 'Rubber price % Change', 'Softlog price % Change', 'Soft sawnwood price % Change', 'Wood pulp price % Change']

# Utilizei a estrutura de LOOP que percorrerá cada elemento para ler e criar um gráfico de histograma com todas as especificações.

for i in range (len(changelist)):
    plt.figure (figsize = (12, 12))
    df [changelist[i]].hist (figsize = (11, 9), linewidth = 1)
    plt.xlabel ('% Change')
    plt.ylabel ('count')
    plt.legend (changelist[i:], loc = 'upper center', bbox_to_anchor = (1.2, 1))
```

A maior parte das matérias-primas tem uma variação ideal até 5%. Ademais, também há matérias-primas mais voláteis que outras.

#### Q2: Encontre a matéria-prima que tem o menor preço ao longo do período analisado.

```python
# Tamanho da área de plotagem:

plt.figure (figsize = (10, 10))

# Criar uma lista das matérias-primas (variáveis) para realizar a correlação e o gráfico:

materialslist = ['Coarse wool Price', 'Copra Price', 'Cotton Price', 'Fine wool Price', 'Hard log Price', 'Hard sawnwood Price', 'Hide Price', 'Plywood Price', 'Rubber Price', 'Softlog Price', 'Soft sawnwood Price', 'Wood pulp Price']

# Utilizei a estrutura de LOOP que percorrerá cada elemento para ler e criar um gráfico com todas as especificações para cada elemento da lista:

for i in range (len(materialslist)):
    plt.subplot (4, 3, i+1)
    plt.subplots_adjust (hspace = 1, wspace = 0.5)
    plt.title (materialslist [i])
    plt.plot (df [materialslist [i]])
    plt.xticks (rotation = 90)
    
plt.suptitle ("Comparação dos preços de matérias-primas")
```

Duas matérias-primas apresentaram um preço mais baixo ao longo do período analisado, são elas: Cotton e Rubber.

#### Q3: Encontre a matéria-prima que tem a menor e maior variação de preço ao longo do período analisado.

```python
# Vou importar a biblioteca RANDOM para randomizar cores dos gráficos:

import random as random
```

```python
# Tamanho da área de plotagem:

plt.figure (figsize = (12, 12))

# Criar uma lista das matérias-primas (variáveis) para realizar a plotagem no gráfico:

changelist = ['Coarse wool price % Change', 'Copra price % Change', 'Cotton price % Change', 'Fine wool price % Change', 'Hard log price % Change', 'Hard sawnwood price % Change', 'Hide price % change', 'Plywood price % Change', 'Rubber price % Change', 'Softlog price % Change', 'Soft sawnwood price % Change', 'Wood pulp price % Change']

# Utilizei a estrutura de LOOP que percorrerá cada elemento para ler e criar um gráfico com todas as especificações para cada elemento da lista:

for i in range (len(changelist)):

    # Vou randomizar as cores dos gráficos:
    r = random.random ()
    g = random.random ()
    b = random.random ()
    color = (r, g, b)
    
    plt.subplot (4, 3, i+1)
    plt.subplots_adjust (hspace = 1, wspace = 0.5)
    plt.title (changelist [i])
    plt.plot (df [changelist [i]], c = color)
    plt.xticks (rotation = 90)
    plt.xlabel ('Years')
    plt.ylabel ('% Change')
    
plt.suptitle ("Comparação dos preços de matérias-primas")
```

### Leitura do gráfico acima e resposta de negócio:

A maior variação de preço se deu na matéria-prima Soft Sawnwood, que ocorreu por volta de 2016. Já a menor variação foi da matéria-prima Plywood.

#### Q4: Encontre as matérias-primas com as mudanças mais drásticas de preço.

```python
# Tamanho da área de plotagem:

plt.figure (figsize = (12, 12))

# Criar uma lista das matérias-primas (variáveis) para realizar a plotagem no gráfico:

pricelist = ['Coarse wool Price', 'Copra Price', 'Cotton Price', 'Fine wool Price', 'Hard log Price', 'Hard sawnwood Price', 'Hide Price', 'Plywood Price', 'Rubber Price', 'Softlog Price', 'Soft sawnwood Price', 'Wood pulp Price']

# Utilizei a estrutura de LOOP que percorrerá cada elemento para ler e criar um gráfico com todas as especificações para cada elemento da lista:

for i in range (len(pricelist)):
    sns.scatterplot (y = df[pricelist[i]], x = df.index)
    plt.legend (pricelist, loc = 'upper center', bbox_to_anchor = (1.2, 1))
    
plt.xlabel ('Years')
plt.ylabel ('Prices')
    
plt.suptitle ("Gráfico de Dispersão dos preços das matérias-primas ao longo do tempo")
```

### Leitura do gráfico acima e resposta de negócio:

De forma geral, as variações mais drásticas aconteceram com a matéria-prima Fine Wool.
